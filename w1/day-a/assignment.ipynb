## Reflection Prompts

- The series is heavily skewed because failure periods drag the left tail down.
- Daily seasonality is weak compared with very strong short-term autocorrelation, so `Rolling IQR` is a reasonable baseline.
- `Isolation Forest` wins because contextual features capture anomaly shape better than a univariate threshold rule.
- Cooldown post-processing improves recall because the NAB labels are anomaly windows, not isolated points.
